1. Write a program which will find all such numbers which are divisible by 7 but are not a
multiple of 5,
between 2000 and 3200 (both included).
The numbers obtained should be printed in a comma-separated sequence on a single
line.
Hints:

2. Write a program which can compute the factorial of a given numbers.
The results should be printed in a comma-separated sequence on a single line.
Suppose the following input is supplied to the program:
8
Then, the output should be:
40320

```python
def fact(x):# Define a function named 'fact' that calculates the factorial of a number 'x'
  if x==0: # Check if the number 'x' is 0
    return 1# If 'x' is 0, return 1 (factorial of 0 is 1)
  return x*fact(x-1)## If 'x' is not 0, recursively call the 'factorial' function with (x-1) and multiply it with 'x'
x=int(input())# Ask the user to input a number to compute its factorial and store it in variable 'x'
print(fact(x))# Print the factorial of the number entered by the user by calling the 'fact' function
```

and 3 (both included) and the values are square of keys.

9. Define a function which can generate a dictionary where the keys are numbers
between 1 and 20 (both included) and the values are square of keys. The function
should just print the values only.
Hints:
Use dict[key]=value pattern to put entry into a dictionary.
Use ** operator to get power of a number.
Use range() for loops.
Use keys() to iterate keys in the dictionary. Also we can use item() to get key/value pairs.

```python
# Create a dictionary with keys as numbers from 1 to 20 and values as their squares
squares_dict = {key: key**2 for key in range(1, 21)}
# Print the values only
print(squares_dict.items())
```
